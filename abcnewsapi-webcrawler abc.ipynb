# NLP Technique - I (News Polarity Analysis based on titles)

```python
titles = list(abcnews_data_final['title'])
count_vectorize = CountVectorizer(lowercase = True, stop_words = 'english', analyzer = 'word').fit(titles)
count_vectorized_corpus = count_vectorize.transform(titles).toarray()
```

```python
word_features = list(count_vectorize.vocabulary_.keys())
word_feature_polarity = np.array([TextBlob(word).sentiment.polarity for word in word_features])
```

```python
word_feature_polarity = word_feature_polarity.reshape(len(word_features), 1)
count_sum_corpus = np.sum(count_vectorized_corpus, axis = 1).reshape(len(titles), 1)
polarity = (count_vectorized_corpus@word_feature_polarity)/count_sum_corpus
```

```python
abcnews_data_final['polarity'] = polarity
abcnews_data_final.head()
```

```python
abcnews_data_final_pos = abcnews_data_final.loc[abcnews_data_final.polarity > 0, :]
abcnews_data_final_neutral = abcnews_data_final.loc[abcnews_data_final.polarity == 0, :]
abcnews_data_final_neg = abcnews_data_final.loc[abcnews_data_final.polarity < 0, :]
```

```python
corpus_pos = list(abcnews_data_final_pos['description'])
for i in range(len(corpus_pos)):
    filtered_list_of_words = [word.lower() for word in corpus_pos[i].split()]
    for j in range(len(filtered_list_of_words)):
        word = ''
        for k in range(len(filtered_list_of_words[j])):
            if filtered_list_of_words[j][k] in string.punctuation:
                word += ''
            else:
                word += filtered_list_of_words[j][k]
        filtered_list_of_words[j] = word
    filtered_list_of_words = [WordNetLemmatizer().lemmatize(word, pos="v") for word in filtered_list_of_words if word not in stopwords.words('english')]
    corpus_pos[i] = ' '.join(filtered_list_of_words)
    
corpus_neg = list(abcnews_data_final_neg['description'])
for i in range(len(corpus_neg)):
    filtered_list_of_words = [word.lower() for word in corpus_neg[i].split()]
    for j in range(len(filtered_list_of_words)):
        word = ''
        for k in range(len(filtered_list_of_words[j])):
            if filtered_list_of_words[j][k] in string.punctuation:
                word += ''
            else:
                word += filtered_list_of_words[j][k]
        filtered_list_of_words[j] = word
    filtered_list_of_words = [WordNetLemmatizer().lemmatize(word, pos="v") for word in filtered_list_of_words if word not in stopwords.words('english')]
    corpus_neg[i] = ' '.join(filtered_list_of_words)
    
corpus_neutral = list(abcnews_data_final_neutral['description'])
for i in range(len(corpus_neutral)):
    filtered_list_of_words = [word.lower() for word in corpus_neutral[i].split()]
    for j in range(len(filtered_list_of_words)):
        word = ''
        for k in range(len(filtered_list_of_words[j])):
            if filtered_list_of_words[j][k] in string.punctuation:
                word += ''
            else:
                word += filtered_list_of_words[j][k]
        filtered_list_of_words[j] = word
    filtered_list_of_words = [WordNetLemmatizer().lemmatize(word, pos="v") for word in filtered_list_of_words if word not in stopwords.words('english')]
    corpus_neutral[i] = ' '.join(filtered_list_of_words)
```

```python
vectorizer_pos = CountVectorizer()
vec_corpus_pos = vectorizer_pos.fit_transform(corpus_pos)
vectorizer_neg = CountVectorizer()
vec_corpus_neg = vectorizer_neg.fit_transform(corpus_neg)
vectorizer_neutral = CountVectorizer()
vec_corpus_neutral = vectorizer_neutral.fit_transform(corpus_neutral)
```
